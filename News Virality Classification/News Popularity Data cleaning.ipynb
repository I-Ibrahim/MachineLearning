```python
import numpy as np
import pandas as pd
```

```python
news=pd.read_csv('OnlineNewsPopularity.csv')
# Feature Selection
news=news.ix[:, ['url',' n_tokens_title',' n_tokens_content',' n_unique_tokens', ' num_hrefs',' num_self_hrefs',
                 ' num_imgs',' num_videos',' average_token_length',
                ' data_channel_is_lifestyle', ' data_channel_is_entertainment',
                ' data_channel_is_bus',' data_channel_is_socmed',
                ' data_channel_is_tech',' data_channel_is_world',' weekday_is_monday',' weekday_is_tuesday',' weekday_is_wednesday',
                ' weekday_is_thursday',' weekday_is_friday',' weekday_is_saturday',' weekday_is_sunday',
                ' is_weekend',' global_rate_positive_words',' global_rate_negative_words',
                ' shares']]

news
```

# Put Days and Category in One Column

```python
day1=[]
day2=[]
day3=[]
day4=[]
day5=[]
day6=[]
day7=[]

for day in news[' weekday_is_monday']:
    if day == 1:
        day1.append('Monday')
    else:
        day1.append(np.nan)
for row in news[' weekday_is_tuesday']:
    if row == 1:
        day2.append('Tuesday')
    else:
        day2.append(np.nan)
for day in news[' weekday_is_wednesday']:
    if day == 1:
        day3.append('Wednesday')
    else:
        day3.append(np.nan)
for day in news[' weekday_is_thursday']:
    if day == 1:
        day4.append('Thursday')
    else:
        day4.append(np.nan)
for day in news[' weekday_is_friday']:
    if day == 1:
        day5.append('Friday')
    else:
        day5.append(np.nan)
for day in news[' weekday_is_saturday']:
    if day == 1:
        day6.append('Saturday')
    else:
        day6.append(np.nan)
for day in news[' weekday_is_sunday']:
    if day == 1:
        day7.append('Sunday')
    else:
        day7.append(np.nan)

        
df=pd.DataFrame({'col1':day1, 'col2':day2, 'col3': day3,'col4':day4,'col5':day5,'col6':day6,'col7':day7})
df=df.fillna('').sum(axis=1)
news['day']=df

news=news.drop({' weekday_is_monday',  ' weekday_is_tuesday',' weekday_is_wednesday',' weekday_is_thursday',
              ' weekday_is_friday',' weekday_is_saturday',' weekday_is_sunday', ' is_weekend'},1)

#change to numerical

news.loc[news['day']=='Monday', 'day']='Monday'
news.loc[news['day']=='Tuesday', 'day']='Tues'
news.loc[news['day']=='Wednesday', 'day']='Wed'
news.loc[news['day']=='Thursday', 'day']=4
news.loc[news['day']=='Friday', 'day']=5
news.loc[news['day']=='Saturday', 'day']=6
news.loc[news['day']=='Sunday', 'day']=7
```

```python
lifestyle=[]
entertainment=[]
business=[]
socialmedia=[]
technology=[]
worldnews=[]

for newstype in news[' data_channel_is_bus']:
    if newstype==1:
        business.append('Business')
    else:
        business.append(np.nan)
for newstype in news[' data_channel_is_entertainment']:
    if newstype==1:
        entertainment.append('Entertainment')
    else:
        entertainment.append(np.nan)
for newstype in news[' data_channel_is_lifestyle']:
    if newstype==1:
        lifestyle.append('Lifestyle')
    else:
        lifestyle.append(np.nan)
for newstype in news[' data_channel_is_socmed']:
    if newstype==1:
        socialmedia.append('Social Media')
    else:
        socialmedia.append(np.nan)
for newstype in news[' data_channel_is_tech']:
    if newstype==1:
        technology.append('Technology')
    else:
        technology.append(np.nan)
for newstype in news[' data_channel_is_world']:
    if newstype==1:
        worldnews.append('World News')
    else:
        worldnews.append(np.nan)
        
        
df2=pd.DataFrame({'col1':business, 'col2':entertainment, 'col3': lifestyle,'col4':socialmedia,
                  'col5':technology,'col6':worldnews})
df2=df2.fillna('').sum(axis=1)
news['Type']=df2

news=news.drop({' data_channel_is_bus', ' data_channel_is_entertainment', ' data_channel_is_lifestyle',
               ' data_channel_is_socmed', ' data_channel_is_tech', ' data_channel_is_world'},1 )


NotTypesum=(news.Type=='Technology').sum()+(news.Type=='World News').sum()+(news.Type=='Lifestyle').sum()+(news.Type=='Entertainment').sum()+(news.Type=='Business')

#fillblank in DF
news=news.replace(r'\s+',np.nan,regex=True).replace('',np.nan)

news.loc[news['Type']=='Technology', 'Type']=1
news.loc[news['Type']=='World News', 'Type']=2
news.loc[news['Type']=='Lifestyle', 'Type']=3
news.loc[news['Type']=='Entertainment', 'Type']=4
news.loc[news['Type']=='Business', 'Type']=5
```
